# To drop unnecessary columns

```python
spoty_data.drop(['id','key','mode','explicit','release_date'], axis=1, inplace=True)
spoty_data.head()
```

# correlation of the audio feature present here


```python
corr = spoty_data[['acousticness', 'danceability','energy','instrumentalness','liveness','tempo','valence','loudness',
                   'speechiness']].corr()

plt.figure(figsize=(20,20))
sns.heatmap(corr, annot=True)
```

# To identify the song trends


```python
y_avg=spoty_data[['acousticness', 'danceability','energy','instrumentalness','liveness','tempo','valence','loudness',
                   'speechiness','year']].\
groupby('year').mean().sort_values(by='year').reset_index()
y_avg.head()
```

```python
plt.figure(figsize=(14,8))
plt.title("Trends of song over course of Time", fontdict={"fontsize":18})
lines=['acousticness', 'danceability','energy','instrumentalness','liveness','valence',
                   'speechiness']
for line in lines:
    ax = sns.lineplot(x='year',y=line,data=y_avg)
plt.ylabel('value')
plt.grid(False)
plt.legend(lines)
```

```python
# To get number of unique artists for further analysis
spoty_data['artists'].nunique()
```

```python
# Top 10 artist with most songs
spoty_data['artists'].value_counts()[0:10]
```

```python
artist_list=spoty_data.artists.value_counts().index[:10]
song_artist = spoty_data[spoty_data.artists.isin(artist_list)][['artists','year']].\
groupby(['artists','year']).size().reset_index(name='song_count')
song_artist.head()
```

```python
plt.figure(figsize=(14,8))
plt.title("No.of songs of every artist as per the year")
sns.lineplot(x='year',y='song_count',hue='artists',data=song_artist)
plt.grid(False)
```

```python
top_artists = pd.DataFrame(np.zeros((100,10)),columns=artist_list)
top_artists['year'] = np.arange(1921,2021)
print(top_artists.shape)
top_artists.head()
```

# Now to convert the long dataframe using melt function

```python
top_artists = top_artists.melt(id_vars='year', var_name='artists', value_name='song_count')
print(top_artists.shape)
top_artists.head()
```

Merge the song counts from song_artist using pandas merge function

```python
merge_data = pd.merge(top_artists, song_artist, on=['year','artists'], how='outer').\
sort_values(by='year').reset_index(drop=True)
merge_data.head()
```

Now in the above output NaN is shown because the artist do not have any particular song in that year, so now replace
NaN values with 0 and drop song_count_x column.

```python
merge_data.fillna(0, inplace=True)
merge_data.drop('song_count_x', axis=1, inplace=True)
merge_data.rename(columns={'song_count_y':'song_count'}, inplace=True)
merge_data.head()
```

To show the cumulative sum of songs that each artist produce over the years.

```python
merge_data['cumsum'] = merge_data[['song_count','artists']].groupby('artists').cumsum()
merge_data.head(20)
```
