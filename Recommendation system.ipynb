```python
import pandas as pd
import numpy as np
import seaborn as sns
import re
import ast
from sklearn.preprocessing import StandardScaler as SS
from sklearn.metrics.pairwise import cosine_similarity as cs
from matplotlib import pyplot as plt
```

```python
def col(df, colname='artists'):
    return np.array([int(x == colname)for x in df.columns]).argmax()
def query_artists(df, lists=[], full=False, strict=True):
    return pd.concat([query_artist(df, string=name, strict=strict)for name in lists], axis=0)
def query_artist(df, string='--', full=False, strict=True):
    lists=[]
    if(not(strict)):
        string = "(.*?)"+string+"(.*?)"
    for i,artist in enumerate(df['artists']):
        if(len(re.findall(string, "".join(artist))) != 0 ):
            if(full):
                lists.append(df.iloc[i])
            else:
                lists.append(df.iloc[i,[col(df,'artists'),col(df,'genres')]])
    if(full):
        return pd.DataFrame(lists, columns=df.columns)
    else:
        return pd.DataFrame(lists, columns=["artists","genres"])

def evaluation(string):
    try:
        return ast.literal_eval(string)
    except:
        return[]

def create_random(df_by_artist, length, score):
    list_of_names = list(set(df_by_artist['artists']))
    random_indices = [round(x) for x in np.random.random(length)*len(list_of_names)]
    random_names = pd.Series(list_of_names).iloc[random_indices].values.tolist()
    random_rates = [int(round(x)) for x in (score[0] + np.random.random(length)*(score[1]-score[0]))]
    name_rate_dict = {}
    for index in range(length):
        name_rate_dict.update({random_names[index]: random_rates[index]})
    return name_rate_dict

def rate_artist(df_by_artist, name_rate_dict):
    name_rate_series = pd.DataFrame({'rate':name_rate_dict.values, 'artists':name_rate_dict.index})
    artists_genre = df_by_artist[df_by_artist['artists'].isin(list(name_rate_dict.keys()))][['artists','genres']]
    df_name_rate = pd.merge(name_rate_series, artists_genre, on='artists', how='inner')
    df_x = df_name_rate.copy()
    for index, genres in enumerate(df_name_rate['genres']):
        for genre in genres:
            df_x.at[index,genre]=1
    df_x=df_x.fillna(0)
    df_user = df_x['rate']
    df_genre_matrix = df_x.drop(['artists','genres','rate'], axis=1).reset_index(drop=True)
    df_profile = df_genre_matrix.transpose().dot(df_user)
    return df_profile

def select_artist(df_by_artist, df_rate):
    list_of_id = []
    for index, row in df_by_artist.iterrows():
        for genre in row['genres']:
            if(genre in df_rate.index):
                list_of_id.append(index)
    list_of_id = list(set(list_of_id))
    df_select_cols = df_by_artist.iloc[list_of_id, [col(df_by_artist,'artists'), col(df_by_artist,'genres')]]
    df_select = df_select_cols.copy()
    for index, row in df_select_cols.iterrows():
        for genre in row['genres']:
            df_select.at[index,genre] = 1
    df_select = df_select.fillna(0)[df_rate.index]
    return df_select

def recommend_artist_w_genre(df_by_artist, name_rate_dict, number):
    df_by_artist = df_by_artist.copy()
    df_by_artist['genres'] = [evaluation(genre) for genre in df_by_artist['genres']]
    name_rate_series = pd.Series(name_rate_dict)
    df_rate = rate_artist(df_by_artist, name_rate_series)
    df_select = select_artist(df_by_artist,df_rate)
    affinity_scores = ((df_select*df_rate).sum(axis=1))/(df_rate.sum())
    affinity_sorting = pd.Series(affinity_scores, name='genre_affinity').sort_values(ascending=False)
    artist_in_df = df_by_artist.iloc[affinity_sorting.index, [col(df_by_artist,'artists')]]
    result_df = pd.concat([affinity_sorting, artist_in_df],axis=1)
    output = result_df[~result_df['artists'].isin(name_rate_series.index)].iloc[:number, :]
    return output.reset_index()

def recommend_artist_w_audio(df_by_artist, name_rate_dict, number):
    artist_given = df_by_artist[df_by_artist['artists'].isin(name_rate_dict.keys())].set_index('artists')[df_by_artist.dtypes[df_by_artist.dtypes == 'float64'].index.tolist()]
    artist = df_by_artist.set_index('artists')[df_by_artist.dtypes[df_by_artist.dtypes=='float64'].index.tolist()]
    name_rate_series = pd.Series(list(name_rate_dict.values()),index = list(name_rate_dict.keys()))
    df_artist_given = pd.DataFrame(SS().fit_transform(artist_given), columns=artist_given.columns, index=artist_given.index)
    df_artist = pd.DataFrame(SS().fit_transform(artist), index=artist.index, columns = artist.columns)
    similar_matrix = pd.DataFrame(cs(df_artist), index=df_artist.index, columns = df_artist.index)[df_artist_given.index]
    top_artist = name_rate_series.dot(similar_matrix.transpose())/name_rate_series.sum()
    top_artist = top_artist[~top_artist.index.isin(name_rate_series.index.tolist())].sort_values(ascending=False)[:number]
    index_top_artist = pd.DataFrame({'artists':df_by_artist[df_by_artist['artists'].isin(top_artist.index)]['artists'].values, 'index':df_by_artist[df_by_artist['artists'].isin(top_artist.index)].index})
    df_top_artist = pd.merge(pd.DataFrame({'audio_affinity':top_artist.values, 'artists':top_artist.index}), index_top_artist, on='artists', how='inner')
    return df_top_artist

def pretty_recommend_artist(df_by_artist, name_rate_dict, number):
    df_genre_affinity = recommend_artist_w_genre(df_by_artist, name_rate_dict, number)
    df_audio_affinity = recommend_artist_w_audio(df_by_artist, name_rate_dict, number)
    print('\n\n---- Genre Affinity ----\n\n')
    for index, row in df_genre_affinity.iterrows():
        print("Number", str(index), ": ", row['artists'],'matching', str(round(row['genre_affinity']*100, 2)), "%")
    print("\n\n--- Audio Affinity ---\n\n")
    for index, row in df_audio_affinity.iterrows():
        print("Number ",str(index),": ",row['artists'],'matching',str(round(row['audio_affinity']*100, 2)), "%")
    print("\n\n")
    df_affinity = pd.merge(df_audio_affinity, df_genre_affinity, on=['index','artists'], how='outer')
    df_scores = pd.concat([pd.DataFrame({'affinity':df_affinity["audio_affinity"].fillna(0)+df_affinity['genre_affinity'].fillna(0)}),df_affinity["artists"]],axis=1)
    plt.figure(figsize=(10,10))
    plt.bar(list(df_scores["artists"]),list(df_scores["affinity"]), color=["green" for x in range(number)] + ["yellow" for x in range(number)])
    plt.xticks(rotation=90)
    plt.xlabel("Artists")
    plt.ylabel("Score")
    plt.title("Audio-Score(Green) vs Genre-Score(Yellow): Comparison")
    return df_affinity
```
